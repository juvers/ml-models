#### Metrics for summarizing model quality
1. Mean Absolute Error (MAE) is a performance metric used to evaluate the 
accuracy of machine learning models, particularly those that involve 
regression problems. It measures the average magnitude of the errors 
between the predicted and actual values. The formula for Mean Absolute 
Error is:

Mean Absolute Error = Σ |y_predicted - y_actual| / n

where 'n' represents the number of data points.

The main advantage of MAE over other error metrics, such as Mean Squared 
Error (MSE), is that it considers only the size of the error rather than 
its squared value. This makes it more suitable for interpreting results 
in cases where the errors are highly variable or have an extreme 
outlier, which may distort the MSE calculation.

2. Mean Squared Error (MSE) is a metric used to evaluate the performance 
of a regression model. It measures the average of the squared difference 
between predicted and actual values in a dataset. The formula for MSE 
is:

Mean Squared Error = Σ (y_predicted - y_actual)^2 / n

where 'n' represents the number of data points.

MSE is widely used due to its simplicity and ease of interpretation. 
However, it can be sensitive to outliers in the data, which may cause 
the metric to overly penalize the model. In such cases, using Root Mean 
Squared Error (RMSE) or Mean Absolute Error (MAE) could be more 
appropriate as they are more robust against outliers.

3. Root Mean Squared Error (RMSE): RMSE measures the square root of the 
mean squared difference between predicted and actual values. This metric 
is often used as an alternative to MSE since it helps normalize the 
errors, making it more robust against outliers.

4. Accuracy: In classification tasks, accuracy is a common metric that 
represents the percentage of correct predictions made by the model. It 
can be calculated using the formula: (TP + TN) / (TP + FP + FN + TN), 
where TP = True Positives, FP = False Positives, FN = False Negatives, 
and TN = True Negatives.

5. Precision and Recall: In classification problems, these metrics are 
used to measure the model's ability to correctly identify positive or 
negative instances. Precision evaluates the proportion of relevant 
instances that are correctly identified (TP / (TP + FP)), while Recall 
measures the proportion of actual positives or negatives that are 
correctly identified (TP / (TP + FN)).

6. F1-score: This metric combines precision and recall using the 
harmonic mean, providing a balanced evaluation for classification 
problems with imbalanced classes. The formula is 2 * ((precision * 
recall) / (precision + recall)).


```python
from sklearn.metrics import mean_absolute_error

# X is a model from the feature matrix
predicted_home_prices = melbourne_model.predict(X)

# y = home prices as outcome variable
mean_absolute_error(y, predicted_home_prices)
```

```python
from sklearn.model_selection import train_test_split

# split data into training and validation data, for both features and target
# The split is based on a random number generator. Supplying a numeric value to
# the random_state argument guarantees we get the same split every time we
# 1. split 2. get model 3. fit with train split set 4. predict using feature matrix 5. run mae on actual outcome vs predicted outcome
train_X, val_X, train_y, val_y = train_test_split(X,y, random_state=0)

melbourne_model = DecisionTreeRegressor()

melbourne_model.fit(train_X, train_y)

val_predictions = melbourne_model.predict(val_X)

print(mean_absolute_error(val_y, val_predictions))
```
